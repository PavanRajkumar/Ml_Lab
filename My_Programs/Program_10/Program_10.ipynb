```python
# https://www.geeksforgeeks.org/ml-locally-weighted-linear-regression/

# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
tou = 0.8
X_train = np.array(list(range(3,33))+[3.2,4.2])
# Convert 1D array to 2D
X_train=X_train[:,np.newaxis]
y_train = np.array([1,2,1,2,1,1,3,4,5,4,5,6,5,6,7,8,9,10,11,11,12,11,11,12,13,16,17,19,18,34,21,22])
```

```python
X_test = np.array([i/10.0 for i in range(400)])
X_test = X_test[:,np.newaxis]
# We need to predict y_test
y_test=[]
```

```python
for r in range(len(X_test)):
    wt = np.exp(-np.sum((X_train-X_test[r])**2,axis=1)/(2*tou)**2)
    w = np.diag(wt)
    fact1 = np.linalg.inv(X_train.T.dot(w).dot(X_train))
    fact2 = X_train.T.dot(w).dot(y_train)
    beta = fact1.dot(fact2)
    prediction = X_test[r].dot(beta)
    y_test.append(prediction)
```

```python
plt.plot(X_train,y_train,'.')
plt.plot(X_test,y_test)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
import sklearn.metrics as sm
from sklearn.metrics import accuracy_score, confusion_matrix
import pandas as pd
import numpy as np
```

```python
iris = datasets.load_iris()

print("Iris Data: \n", iris.data[:10])
print("\n\nIris Features: \n", iris.feature_names)
print("\n\nIris Target: \n", iris.target)
print("\n\nIris Target Names: \n", iris.target_names)
```

```python
X = pd.DataFrame(iris.data)
print(X)
```

```python
# View the distribution
X.columns = ['Sepal_Length', 'Sepal_Width','Petal_Length', 'Petal_Width']
y = pd.DataFrame(iris.target)
y.columns = ['Targets']

plt.figure(figsize = (14,7))
colormap = np.array(['red','lime','black'])

plt.subplot(1,2,1)
plt.scatter(X.Sepal_Length, X.Sepal_Width, c = colormap[y.Targets], s=40)
plt.title('Sepal');


plt.subplot(1,2,2)
plt.scatter(X.Petal_Length, X.Petal_Width, c = colormap[y.Targets], s=40)
plt.title('Petal');
```

```python
# Build KMeans model
model = KMeans(n_clusters=3)
model.fit(X) # Unsupervised learning thus, passing in only the features to the model
```

```python
pred_y = model.labels_
print(pred_y)
```

```python
sm.accuracy_score(y, pred_y)
```

```python
sm.confusion_matrix(y, pred_y)
```

```python
# View the results
plt.figure(figsize = (14,7))

colormap = np.array(['orange', 'blue', 'grey'])

plt.subplot(1,2,1)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y.Targets], s=40)
plt.title("Real Classification");

plt.subplot(1,2,2)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[pred_y], s=40)
plt.title("KMeans Classification");
```

```python
# For EM Algorithm
from sklearn import preprocessing 

scaler = preprocessing.StandardScaler() #Performs normalisation of the data
scaler.fit(X)
xsa = scaler.fit_transform(X)
xs = pd.DataFrame(xsa, columns=X.columns)
xs.sample(5)
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=3) # number of 3 is passed because there are 3 target classes 
gmm.fit(xs)
```

```python
y_cluster_gmm = gmm.predict(xs)
```

```python
plt.subplot(1,2,1)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y.Targets], s=40)
plt.title("Real Classification")

plt.subplot(1,2,2)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y_cluster_gmm], s=40)
plt.title('GMM Classification')

```

```python
sm.accuracy_score(y,y_cluster_gmm)
```

```python
sm.confusion_matrix(y, y_cluster_gmm)
```

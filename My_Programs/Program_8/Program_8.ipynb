```python
import matplotlib.pyplot as plt
from sklearn import datasets, preprocessing
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.mixture import GaussianMixture
import pandas as pd
```

```python
iris = datasets.load_iris()
X = pd.DataFrame(iris.data)
X.columns = ['Sepal_Length','Sepal_Width','Petal_Length', 'Petal_Width']
y = pd.DataFrame(iris.target)
y.columns = ['Targets']
```

```python
model = KMeans(n_clusters=3)
# Unsupervised learning thus, passing in only the features to the model
model.fit(X) 
y_pred = model.labels_
print(accuracy_score(y, y_pred))
```

```python
# For EM Algorithm
scaler = preprocessing.StandardScaler() #Performs normalisation of the data
scaler.fit(X)
xsa = scaler.fit_transform(X)
xs = pd.DataFrame(xsa)
gmm = GaussianMixture(n_components=3) # number of 3 is passed because there are 3 target classes 
gmm.fit(xs)
y_gmm = gmm.predict(xs)
print(accuracy_score(y,y_gmm))
```
